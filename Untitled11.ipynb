```python
from bs4 import BeautifulSoup
import requests
import locale
```

```python
url="https://indianexpress.com/section/news-archive/"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
#print(soup.prettify()) # print the parsed data of html
```

```python
pageCount = ''

for foo in soup.find_all('div', attrs={'class': 'leftpanel'}):
    foo_descendants = foo.descendants
    for d in foo_descendants:
        if d.name == 'div' and d.get('class', '') == ['nation']:
          print(d.get('class', '')) 
          pageTag = d.find("div", {"class": "pagination"}) 
          ultag = pageTag.find("ul", {"class": "page-numbers"})
          #count = len(list(ultag.descendants))  
          lilist = ultag.find_all('li')  
          count= len(lilist)
          print(lilist[count-2].text)
          pageCount = lilist[count-2].text          
```

```python
#https://indianexpress.com/section/news-archive/page/2/
print(pageCount)
locale.setlocale(locale.LC_ALL, 'en_US.UTF8')
print(locale.atoi(pageCount))
for pageSize in range(1, 100):
  url="https://indianexpress.com/section/news-archive/page/"+ str(pageSize) + "/"

  # Make a GET request to fetch the raw HTML content
  html_content = requests.get(url).text

  # Parse the html content
  soup = BeautifulSoup(html_content, "lxml")
  for foo in soup.find_all('div', attrs={'class': 'leftpanel'}):
    foo_descendants = foo.descendants
    for d in foo_descendants:
        if d.name == 'div' and d.get('class', '') == ['nation']:
          print(' PAGE  NO :::: ', pageSize)           
          tdTags = d.find_all("div", {"class": "articles"})           
          for tag in tdTags:
              title = tag.find("h2", {"class": "title"})
              link = title.find("a")
              print (link.get("href"))
              date = tag.find("div", {"class": "date"})
              print(date.text)
              para = tag.find("p")
              print(para.text)  
```
